# Using pandas reading a db

```python
conection = sqlite3.connect(
    'RDP-Reading-Data-with-Python-and-Pandas-master/unit-1-reading-data-with-python-and-pandas/lesson-11-reading-data-from-relational-databases/files/chinook.db'
)
```

```python
df = pd.read_sql('select * from employees;',conection)
```

```python
df 
```

```python
df = pd.read_sql('select * from employees;', conection,
                index_col='EmployeeId',
                parse_dates=['BirthDate', 'HireDate']
)
```

```python
df
```

```python
df.dtypes
```

## Using read_sql_query

```python
df = pd.read_sql_query('select * from employees limit 5', conection)
```

```python
df 
```

```python
df = pd.read_sql_query('select * from employees;', conection,
                      index_col='EmployeeId',
                      parse_dates=['BirthDate', 'HireDate'])
```

```python
df 
```

### From a DF to a table

```python
df.to_sql('employees222', conection)
```

```python
pd.read_sql_query('select * from employees222;', conection)
```

```python
conection.close()
```
